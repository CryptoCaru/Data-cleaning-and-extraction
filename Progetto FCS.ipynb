<span style="color:green;font-weight:bold;font-style:italic;font-size:14px">DATA CLEANING </span>

```python
gps = gps.drop(gps.index[10472])
gps[gps == 'Varies with device'] = np.nan

gpsur = gpsur.dropna()

gps = gps.reset_index(drop=True)
gpsur = gpsur.reset_index(drop=True)
```

<span style="color:green;font-weight:bold;font-size:20px"> EXERCISE 1: Convert the app sizes to a number

```python
measure = gps['Size']
measure = measure.replace(['[0-9][.]','M','k'],['','1000000','1000'],regex=True)
gps['Size'] = pd.to_numeric(gps['Size'].replace(["M","k"],['',''],regex=True), errors='ignore') # errors to ignore NaN
measure = np.array(measure, dtype=float) # converting to float and not to int because there are NaN
gps['Size'] = gps['Size']*measure
```

<span style="color:green;font-weight:bold;font-size:20px"> EXERCISE 2: Convert the number of installs to a number

```python
gps['Installs'] = gps['Installs'].replace([',','[+]'],['',''],regex=True)
gps['Installs'] = pd.to_numeric(gps['Installs'])
```

```python
gps.head()
```

<span style="color:green;font-weight:bold;font-size:20px">EXERCISE 3: Transform “Varies with device” into a missing value

```python
gps[gps == 'Varies with device'] = np.nan
```

```python
gps[gps == "Varies with device"].count()
```

<span style="color:green;font-weight:bold;font-size:20px"> EXERCISE 4: Convert Current Ver and Android Ver into a dotted number (e.g. 4.0.3 or 4.2)

```python
# remove if starts or ends with "." or is not 0-9 or "."
gps['Current Ver'] = gps['Current Ver'].replace(["^[\.]","[^\d\.]","[\.]$"], ["", "", ""], regex=True)
gps['Android Ver'] = gps['Android Ver'].replace(["^[\.]","[^\d\.]","[\.]$"], ["", "", ""], regex=True)
gps = gps.replace(r'^\s*$', np.nan, regex=True) # fill empty cells with NaN
```

<span style="color:green;font-weight:bold;font-size:20px"> EXERCISE 5: Remove the duplicates

```python
gps.drop_duplicates(inplace = True)
gps = gps.reset_index(drop=True)
```

<span style="color:green;font-weight:bold;font-size:20px"> EXERCISE 6: For each category, compute the number of apps

```python
Categories_appl = gps.groupby(['Category'])[['App']].count()
```

<span style="color:green;font-weight:bold;font-size:20px"> EXERCISE 7: For each category, compute the average rating

```python
Categories_rating = gps.groupby(['Category'])[['Rating']].mean()
```

<span style="color:green;font-weight:bold;font-size:20px"> EXERCISE 8: Create two dataframes: one for the genres and one bridging apps and genres. So that, for instance, the app Pixel Draw - Number Art Coloring Book appears twice in the bridging table, once for Art & Design, once for Creativity

```python
# split genres 
dupl = gps['Genres'].str.split(";", n = 1, expand = True) 

gps_1 = gps.copy()
gps_2 = gps.copy()

gps_1['Genres'] = dupl[0]

# remove null rows
gps_2['Genres'] = dupl[1]
gps_2 = gps_2[gps_2['Genres'].notnull()]

# append in the end of the df the rows with single genre
gps = pd.concat([gps_1,gps_2])
gps = gps.reset_index(drop=True)

# find genres
Genres = gps['Genres'].drop_duplicates()
Genres = pd.DataFrame({'Genres':Genres})
Genres = Genres.reset_index(drop=True)

# table bridging apps and genres
AppGen = gps[['App','Genres']]
```

<span style="color:green;font-weight:bold;font-size:20px"> EXERCISE 9: For each genre, create a new column of the original dataframe. The new columns must have boolean values (True if the app has a given genre)

```python
Header = Genres['Genres']

Genres = pd.DataFrame(index = range(0,len(gps)), columns=Header)

for i in range(0,len(Genres.columns)):
    Genres[Genres.columns[i]]=Genres.columns[i]

for i in range(0,len(Genres.columns)):
    Bool = Genres[Genres.columns[i]] == gps['Genres']
    Genres[Genres.columns[i]] = Bool
    
gps = pd.concat([gps,Genres], axis=1, sort=False)
```

```python
gps.head()
```

<span style="color:green;font-weight:bold;font-size:20px">EXERCISE 10: For each genre, compute the average rating. What is the genre with highest average?

```python
Genres_2 = gps.groupby(['Genres']).mean()
Genres_2 = Genres_2['Rating']
Max = Genres_2[Genres_2 == Genres_2.max()]
print('The genre with maximum average is', Max.index[0], 'with AVG', round(Max[0],2))
```

<span style="color:green;font-weight:bold;font-size:20px"> EXERCISE 11: For each app, compute the approximate income, obtain as a product of number of installs and price.

```python
gps['Price'] = pd.to_numeric(gps['Price'].replace('[$]','', regex=True))
gps['Approximate income'] = gps['Price']*gps['Installs']
```
